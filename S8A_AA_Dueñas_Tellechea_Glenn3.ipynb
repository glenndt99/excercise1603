Algoritomos de clasificacion SVM, Navie Bayes, Nearest Neighbors y Decision Trees aplicados al dataset de Iris

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import svm, tree, metrics
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
features = iris.data
X = np.array(features)
y = np.array(iris.target)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.20)
```

#### SVM

```python
classifer = svm.SVC()
classifer.fit(X_train,y_train)
print("Exactitud = ",classifer.score(X_test,y_test))
```

#### Naive Bayes

```python
nb = GaussianNB()
nb.fit(X_train,y_train).predict(X_test)
print("Exactitud = ", nb.score(X_test, y_test))
```

#### Nearest Neighbors

```python
nn = KNeighborsClassifier(n_neighbors=3,p=2, metric='minkowski')
nn.fit(X_train,y_train)
print("Exactitud = ", nn.score(X_test,y_test))
```

#### Decision Tree

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X_train,y_train)
print("Exactitud = ", clf.score(X_test,y_test))
```
